1.3 Напишите функцию, которая по `username` ищет человека и выводит информацию о нем. Если пользователь с заданным `username` отсутствует, возбудите исключение `ValueError`

````python
def search(username):
    if username not in [i['username'] for i in json_data]:
        raise ValueError('Нет такого пользователя!')
    for i in json_data:
        if i['username']==username:
            return i
            
search(input('Введите имя\n'))
````

````python
import pickle
````

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
raw_jobs = [i['jobs'] for i in json_data]
usernames = [i['username'] for i in json_data]
user_raw_jobs_dic = {k:v for k,v in zip(usernames, raw_jobs)}
jobs=set()
for list_of_jobs in raw_jobs:
    jobs.update(list_of_jobs)
jobs_username_dic = {}
for job in jobs:
    users_with_the_job = [username for username in usernames if job in user_raw_jobs_dic[username]]
    jobs_username_dic[job]=users_with_the_job
jobs_username_dic
````

````python
jobs
````

````python
post = {i for i in sum([i['jobs'] for i in json_data], [])}
posts_people = {k:[v['username'] for v in json_data if k in v['jobs'] ] for k in post}
posts_people
````

````python
with open('job_people.pickle', 'wb') as pickle_file:
    pickle.dump(jobs_username_dic, pickle_file)

with open('job_people.json', 'w') as json_file:
    json.dump(jobs_username_dic, json_file)
````

### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
import json
````

````python
import requests
from bs4 import BeautifulSoup
with open('data/steps_sample.xml') as xml_file:
    content = BeautifulSoup(xml_file, "xml")
````

````python
content
````

````python
list_of_recipes = content.recipes.find_all('recipe')
list_of_recipes
````

````python
id_steps_dic = {}
for i in list_of_recipes:
    id_steps_dic[i.id.text] = [step.text.replace('\n', '') for step in i.steps if step.text!='\n']

with open('id_steps_dic.json', 'w') as result_dic:
    json.dump(id_steps_dic, result_dic)
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
list_of_recipes = content.recipes.find_all('recipe')
id_len_steps={i.id.text:len([step.text.replace('\n','') for step in i.steps if step!='\n']) for i in list_of_recipes}
lengths_of_steps = set(id_len_steps.values())
res_dic={}
for length in lengths_of_steps:
    corresponding_ids=[]
    for pair in list(id_len_steps.items()):
        if pair[1]==length:
            corresponding_ids.append(pair[0])
    res_dic[length] = corresponding_ids
res_dic
````

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

````python
list_of_recipes = content.recipes.find_all('recipe')
result=[i.id.text for i in list_of_recipes if i.steps.find_all(attrs={'has_minutes':'1'}) or i.steps.find_all(attrs={'has_hours':'1'})]
print(result[:5])
````
