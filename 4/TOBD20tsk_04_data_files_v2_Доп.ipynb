```python
import urllib
import urllib.parse
import urllib.request

import requests
from bs4 import BeautifulSoup
```

```python
print("hi")
```

```python
wiki_hom = 'https://ru.wikipedia.org'
url_wiki = 'https://ru.wikipedia.org/wiki/Категория:Города_по_субъектам_Российской_Федерации'
url_wiki_q  = urllib.parse.quote(url_wiki, safe='/:')
```

```python
page = BeautifulSoup(urllib.request.urlopen(url_wiki_q), 'lxml')
link_to_county_towns = []
for mw_category_group in page.find_all('div', class_='mw-category-group'):
    if mw_category_group.find("h3").text == "С":
        for link in mw_category_group.find_all("a"):
            link_to_county_towns.append((link["href"], link.text))

subcategories = {}

for (link, text) in link_to_county_towns:
    page = BeautifulSoup(urllib.request.urlopen(wiki_hom+link), 'lxml')
    subcategories[text] = []
    list_a = subcategories[text]
    page = page.find('div', class_='mw-content-ltr')
    for subcat in page.find_all('div', class_='mw-category-group'):
        for a in subcat.find_all("a"):
            list_a.append((a.text, a["href"]))
```
